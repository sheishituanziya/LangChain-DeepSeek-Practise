## 三、API密钥安全管理

### 1. 密钥获取流程

DeepSeek API密钥的获取步骤包括：
1. 访问DeepSeek官方网站
2. 注册开发者账号
3. 创建新应用
4. 获取专属API密钥

### 2. 安全存储方案

**绝对避免**将API密钥直接硬编码在源代码中！推荐的安全实践是：

1. 创建`.env`文件存储敏感信息：
```
DEEPSEEK_API_KEY=your_actual_api_key
DEEPSEEK_API_BASE=https://api.deepseek.com/v1
```

2. 使用`python-dotenv`库安全加载：

````python
from dotenv import load_dotenv, find_dotenv

def load_deepseek_config():
    """安全加载DeepSeek API配置"""
    # 自动查找并加载.env文件
    load_dotenv(find_dotenv())

load_deepseek_config()
````

这种方案的优势：
- 密钥不会进入版本控制系统
- 方便不同环境配置切换
- 符合安全开发最佳实践

## 四、DeepSeek客户端初始化

### 1. 客户端配置

通过LangChain的DeepSeek接口，可以轻松创建客户端实例：

````python
from langchain_deepseek import ChatDeepSeek

llm = ChatDeepSeek(
    model="deepseek-chat",  # 指定模型版本
    max_retries=2,          # 网络异常自动重试
    temperature=0.7,        # 控制生成随机性
    timeout=30              # 请求超时设置(秒)
)
````

### 2. 核心参数说明

| 参数名 | 类型 | 默认值 | 说明 |
|--------|------|--------|------|
| model | str | 必填 | 模型版本标识 |
| max_retries | int | 2 | 失败自动重试次数 |
| temperature | float | 0.7 | 生成多样性控制(0-1) |
| timeout | int | 30 | 请求超时时间 |

## 五、基础API调用实践

### 1. 简单问答实现

定义标准化的对话函数：

````python
def get_completion(text):
    messages = [
        ("system", "你是Ai助手，可以帮助我回答生活中的各种问题"),
        ("human", text),
    ]
    response = llm.invoke(messages)
    return response.content
````

### 2. 实际调用示例

````python
query = "用200字左右介绍人工智能的现状"
print(get_completion(query))
````
